Замечаем, что перввый текст и третий тексты, преимущественно, состоит из иероглифов. Необходимо понять, чем отличаются корейские, японские, китайские иероглифы. 
1. В корейском языке используется алфавит Хангыль, отличительная черта которого обилие кругов, овалов и прямых линий.
2. В японском языке три основных элемента: хигарана, катакана и кандзи. Хигарана и катакана – это слоговые системы, а кандзи происходит от китайский иероглифов. Большинство иероглифов хирагана округлые, но не такой аккуратной формы, как корейские . В катакана, с другой стороны, используются в основном прямые или чуть округлые линии, составляющие довольно простые комбинации . Ни в китайском, ни в корейском эти системы не используются.Соответственно, присутсвие в тексте таких элементов позволяет определить японский язык.
3. Если нет характерных форм корейских иероглифов Хангыль или японских иероглифов хирагана или катакана, скорее всего, это  китайский язык. Китайская письменность состоит их сложных иероглифов, которые называются хандзы (по-японски – кандзи, по-кореский – ханджа). Все эти иероглифы могут встретиться и в японском языке, однако лишь в сочетании с хирагана и катакана. 

По кодам символов(в таблице unicode) понимаем, что:
* Первый текст - китайский язык.(нет хираганы или катаканы, обилия кругов, коды символов в основном в таком интервале(0x4E00—0x9FFF), который соответствует  Унифицированным иероглифам ККЯ) 
* Второй текст - английский язык.(Латинница - (0x0020—0x007F))
* Третий текст - японский язык.(Катакана и Хирагана - (0x3040—0x30FF))



Имеем три метки: китайский, японский, английский. Теперь несложно понять, как ,исходя из кодов символов, понять, какому языку соотвествует строка. По таблице unicode определим следующие группы:
* Управляющие символы (0000—001F) + цифры (0030 - 0039). 
Относятся ко всем трем языкам, значит никак не влияют на классифицирование строк.
* Латинница(0020—007F) + знаки препинания(2000—206F).
Относим к английскому.  Данная группа знаков препинания тоже относится сюда, тк в японском и китайском другие знаки препинания.
* Символы и пунктуация ККЯ(3000—303F) +  Полуширинные и полноширинные формы(FF00—FFEF)- общая группа для китайского и японского языков.
* Унифицированные иероглифы ККЯ(4E00—9FFF)
Относим к китайскому, т.к. в японском обязательно должны быть Катакана и Хирагана.
* Катакана и Хирагана(3040-30FF)
Относим к японскому.

Алгоритм работы классификаторы: определяем по каждому символу строки к какому языку он принадлежит(по коду), выбираем самый часто встречающийся язык.

```python
from collections import Counter
class StringClassifier:
    def predict(self, text):
        """
        Input:
        text of strings to classify.
        Output:
        list of labels.
        """
        labels = []
        for str in text:
            languages = Counter()
            # 0 - английский.
            # 1 - японский.
            # 2 - китайский.
            for symb in str:
                code = ord(symb)
                #Проверяем первую группу - управляющий или цифра.
                if code <= 0x0020 or 0x0030 <= code <= 0x0039:
                    continue
                #Проверяем вторую группу - английский.
                elif 0x0020 <= code <= 0x007F or 0x2000 <= code <= 0x206F:
                    languages[0] += 1
                #Проверяем Катакана и Хирагана, что однозначно определяет японский.
                elif 0x3040 <= code <= 0x30FF:
                    languages[1] += 1
                #Проверяем унифицированные иероглифы ККЯ, что определяет китайский.
                elif 0x4E00 <= code <= 0x9FFF:
                    languages[2] += 1
                #Проверяем символы и пунктуацию ККЯ + полуширинные и полноширинные формы.
                elif 0x3000 <= code <= 0x303F or 0xFF00 <= code <= 0xFFEF:
                    languages[1] += 1
                    languages[2] += 1
            cat = languages.most_common(1)
            if len(cat) != 0:
                labels.append(cat[0][0])
            else:
                labels.append(0)
        return labels        
    def accuracy(self, labels, predict):
        """
        input:
        list of real labels, list of predictions.
        output:
        float: labels of classifier.
        """
        right = 0
        for i in range(len(labels)):
            if labels[i] == predict[i]:
                right += 1
        acc = right / len(labels)
        return acc
```

Проверяем классификатор на данных текстах.

```python
text = np.array(text_lines_3 + text_lines_4 + text_lines_2)
k_fold_list = list(KFold(len(text), n_folds=10, shuffle=True))
labels = np.array([0]*len(text_lines_3) + [1]*len(text_lines_4) + [2]*len(text_lines_2))
```

```python
accuracies = []
for it_train, it_test in k_fold_list:
    x_test = text[it_test]
    y_test = labels[it_test]


    classifier = StringClassifier()

    prediction = classifier.predict(x_test)
    
    accuracy = accuracy_score(prediction, y_test)
    accuracies.append(accuracy)
print(accuracies)
```

```python
#Получаем строку, по которой будем кодировать. Также храним индексы символов.
index = {}
s = string.ascii_letters
s1 = s[26:]
s2 = s[:26]
s = s1 + s2
s = s + string.digits 
s = s + '+/'
for i in range(len(s)):
    index[s[i]] = i
```

```python
def encode(st, s):
    n_st = (16 - len(bin(ord(st[0]))[2:])) * '0' + bin(ord(st[0]))[2: ]
    n_st = n_st + (16 - len(bin(ord(st[1]))[2:]))*'0' + bin(ord(st[1]))[2: ]
    n_st = n_st + (16 - len(bin(ord(st[2]))[2:]))*'0' + bin(ord(st[2]))[2: ]
    code_st = ''
    k = 48
    while k > 0:
        i = int(n_st[:6],2)
        code_st = code_st + s[i]
        n_st = n_st[6:]
        k -= 6
    return code_st
```

```python
#Функция кодирования.
def encoding(text, s):
    """
    input : text to code.
   
    output: code of the text by the line 
    "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789+/".
    """
    if len(text) == 0:
        return ''
    elif len(text) < 3:
        if len(text) == 1:
            text = text + '\x00\x00'
            code_st = encode(text, s) + '=='
            return code_st
        if len(text) == 2:
            text = text + '\x00'
            code_st = encode(text, s) + '='
            return code_st
    else: 
        r = len(text) % 3
        if r == 1:
            text = text + '\x00\x00'
        elif r == 2: 
            text = text + '\x00'
        tmp = len(text)
        code_st = ''
        new_text = text
        while tmp > 0:
            code_st = code_st + encode(new_text[:3], s)
            new_text = new_text[3:]
            tmp -= 3
        code_st = code_st + '=' * r
        return code_st    
```

```python
def decode(code, index):
    n_st = ''
    st = ''
    for i in range(8):
        n_st = n_st + (6 - len(bin(index[code[i]])[2:]))*'0' + bin(index[code[i]])[2:]
    for i in range(3):
        st = st + chr(int(n_st[:16],2))
        n_st = n_st[16:]
    return st
```

```python
def decoding(code, index):
    if len(code) == 0:
        return ''
    r = 0
    if code[-1] == '=':
        r = 1
    if code[-2] == '=':
        r = 2
    if r != 0:
        code = code[:-r]
    tmp = len(code)
    st = ''
    new_code = code
    while tmp > 0:
        st = st + decode(new_code[:8], index)
        new_code = new_code[8:]
        tmp -= 8
    if r != 0:
        st = st[:-r]
    return st
```
